```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import *
import seaborn as sns
from matplotlib import rcParams
import datetime as dt
import glob, os  
```

```python
# merge all the monthly csv files from 2018
path = r'C:\Users\jjtre\Documents\UCSD Coding Bootcamp Data Science\Class Notes\tableau\Homework\Data\Bike Trip Histories\2018'                    
all_files = glob.glob(os.path.join(path, "*.csv"))     

df_from_each_file = (pd.read_csv(f) for f in all_files)
concatenated_2018_df = pd.concat(df_from_each_file, ignore_index=True)
concatenated_2018_df.head()
```

```python
# check for missing records
pd.concat([concatenated_2018_df.isnull().sum(), 100 * concatenated_2018_df.isnull().sum()/len(concatenated_2018_df)], 
              axis=1).rename(columns={0:'Missing Records', 1:'Percentage (%)'})
```

```python
# rename column headers
concatenated_2018_df.rename(columns={'tripduration': 'Trip Duration (Sec)', 'starttime': 'Start Time', 'stoptime': 'Stop Time', 'start station id': 'Start Station ID', 'start station name': 'Start Station Name', 'start station latitude': 'Start Station Latitude', 'start station longitude': 'Start Station Longitude', 'end station id': 'End Station ID', 'end station name': 'End Station Name', 'end station latitude': 'End Station Latitude', 'end station longitude': 'End Station Longitude', 'bikeid': 'Bike ID', 'usertype': 'User Type', 'birth year': 'Birth Year', 'gender': 'Gender'}, inplace=True)
concatenated_2018_df.head()
```

```python
broken_bike_df = pd.DataFrame()
broken_bike_df['Less than 90sec same station'] = concatenated_2018_df[(concatenated_2018_df['Trip Duration (Sec)'] < 90) & (concatenated_2018_df['Start Station ID'] == concatenated_2018_df['End Station ID'])]['Bike ID']
broken_bike_df.head()                                                                   
```

```python
broken_bike_df.to_csv("broken_bike.csv", encoding='utf-8', index=False)
```

```python
# check the number of rides that were less than 90sec and start and stop at the same station (errors, false starts, redocking, tests)
check_false_starts_df = pd.DataFrame()
check_false_starts_df['First Bike'] = concatenated_2018_df[concatenated_2018_df['Trip Duration (Sec)'] < 90]['Start Station Name'] 
check_false_starts_df['Second Bike'] = concatenated_2018_df[concatenated_2018_df['Trip Duration (Sec)'] < 90]['End Station Name']
check_false_starts_df.head()
```

```python
# 45 mins ride duration plus $2.50/ 15 mins after that
ride_duration_too_long_df = pd.DataFrame()
ride_duration_too_long_df['Longer Than 2 hours'] = concatenated_2018_df[concatenated_2018_df['Trip Duration (Sec)'] > 7200]['Start Station Name'] 
ride_duration_too_long_df['Longer Than 4 hours'] = concatenated_2018_df[concatenated_2018_df['Trip Duration (Sec)'] > 14400]['Start Station Name'] 
ride_duration_too_long_df['Longer Than 8 hours'] = concatenated_2018_df[concatenated_2018_df['Trip Duration (Sec)'] > 28800]['Start Station Name']
ride_duration_too_long_df.head()
```

```python
ride_duration_too_long_df.count()
```

```python
drop_false_starts_df = concatenated_2018_df.drop(concatenated_2018_df.index[(concatenated_2018_df['Trip Duration (Sec)'] < 90) & (concatenated_2018_df['Start Station ID'] == concatenated_2018_df['End Station ID'])])
drop_false_starts_df.head()
```

```python
drop_false_starts_df.describe()
```

```python
drop_greater_than_4hr_df = drop_false_starts_df.drop(drop_false_starts_df.index[(drop_false_starts_df['Trip Duration (Sec)'] > 14400)])
drop_greater_than_4hr_df.head()
```

```python
drop_greater_than_4hr_df.describe()
```

```python
drop_greater_than_4hr_df['Age'] = (2018 - drop_greater_than_4hr_df["Birth Year"])
drop_greater_than_4hr_df.head()
```

```python
drop_greater_than_4hr_df.dtypes
```

```python
too_old_df = pd.DataFrame()
too_old_df['Older than 90'] = drop_greater_than_4hr_df[drop_greater_than_4hr_df['Age'] > 90]["Birth Year"]
too_old_df['Older than 80'] = drop_greater_than_4hr_df[drop_greater_than_4hr_df['Age'] > 80]["Birth Year"]
too_old_df['Older than 70'] = drop_greater_than_4hr_df[drop_greater_than_4hr_df['Age'] > 70]["Birth Year"]
too_old_df.head()
```

```python
too_old_df.describe()
```

```python
# drop anyone riding a bike older than 90.  lying or error on age = 130 rows
drop_older_90_df = drop_greater_than_4hr_df.drop(drop_greater_than_4hr_df.index[(drop_greater_than_4hr_df['Birth Year'] < 1928)])
drop_older_90_df.head()
```

```python
drop_older_90_df.describe()
```

```python
drop_older_90_df.to_csv("cleaned_citybike_tripdata_2018.csv", encoding='utf-8', index=False)
```
